```python
import gensim
from gensim.models import FastText 
```

```python
# import corpus and 
from gensim.models.word2vec import LineSentence
data = LineSentence('Output.txt')
```

```python
#model = ft.load_fasttext_format("crawl-300d-2M.vec") 
model = FastText(data, size=300, window=3, min_count=1, iter=10)
```

```python
from gensim.test.utils import get_tmpfile

fname = get_tmpfile("fasttext.model")
#model.save(fname)
```

```python
model = FastText.load(fname)  # loading model
```
